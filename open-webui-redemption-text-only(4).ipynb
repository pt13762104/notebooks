```python
#https://stackoverflow.com/questions/1196074/start-a-background-process-in-python
import os
os.environ["OLLAMA_FLASH_ATTENTION"]="1"
os.environ["WEBUI_AUTH"]="0"
os.environ["LD_LIBRARY_PATH"]="/kaggle/working/openwebui/lib64/python3.11/site-packages/nvidia/nvjitlink/lib:"+os.environ["LD_LIBRARY_PATH"]
import subprocess
```

```python
def run_in_background(s):
    subprocess.Popen(s.split())
```
